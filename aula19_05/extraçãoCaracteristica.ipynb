```python
import cv2 as cv
import numpy as np

imagem = cv.imread("quadrado.png",0)
limiar, imgBinarizada = cv.threshold(imagem, 127,255, cv.THRESH_BINARY)

contornos, hierarquia = cv.findContours(imgBinarizada, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLES)

print(contornos[0])

area = cv.contourArea(contornos[0])

cv.waitKey(0)
cv.destroyAllWindows()
```
